#### **5.5. Machine Learning Models**

- **Logistic Regression (LR)**:
  Logistic regression is a classification algorithm that models the probability of a binary outcome based on one or more predictor variables using a logistic function.
  
- **Support Vector Machines (SVM)**:
  SVM finds a hyperplane that best separates the data into classes by maximizing the margin between the closest points of the classes (support vectors).

- **Stochastic Gradient Descent (SGD)**:
  An iterative optimization algorithm used for minimizing loss functions in linear classifiers and regressors, particularly useful for large datasets.

- **Decision Trees**:
  A tree-based algorithm where each internal node represents a decision on a feature, each leaf node represents an outcome, and paths from root to leaf represent classification rules.

- **K-Nearest Neighbors (KNN)**:
  A lazy learning algorithm that classifies a data point based on the majority class of its k-nearest neighbors in the feature space.
  
- **Naive Bayes (NB)**:
  A probabilistic classifier based on Bayes' theorem, assuming strong independence between features.

- **Bagging (Bootstrap Aggregating)**:
  Combines predictions from multiple models (e.g., Random Forest) to reduce variance and improve accuracy.

- **Boosting**:
  Combines weak learners sequentially to form a strong learner by minimizing errors iteratively (e.g., Extreme Gradient Boosting - XGBoost).

  - Specificity: True Negative Rate
  - Accuracy: Overall correctness
  - Precision: Positive Predictive Value
  - F1-Score: Harmonic mean of Precision and Recall
  - AUC: Area Under the ROC Curve

- **K-Fold Cross-Validation**: Splits data into k subsets; trains on k-1 and tests on the remaining one iteratively to ensure robust performance evaluation.

```python
# Confidence Interval for the Mean
def confidence_interval(data, confidence=0.95):
    mean = data.mean()
    std = data.std()
    n = len(data)
    z = stats.norm.ppf(1 - (1 - confidence) / 2)  # z-score for 95% CI
    margin_of_error = z * (std / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return (lower_bound, upper_bound)

ci_results = {}
for col in vr56_non_cat_vars:
    ci_results[col] = confidence_interval(vr56_sample[col])


print("\nConfidence Intervals (95%):")
for col, ci in ci_results.items():
    print(f"{col}: {ci[0]:.2f} to {ci[1]:.2f}")
```

####**> Observations:**

Centralized Distribution: The narrow intervals, centered around 0.5, strongly suggest that the true population means for Quantity, Value, and Weight are likely very close to 0.5.

Comparable Uncertainty: The similar range of the confidence intervals indicates comparable levels of uncertainty in estimating the population means for all three variables.

## **Inferential Statistics**

```python
# Test of Homogeneity {Chi-sq} on categorical data

vr56_contingency_table = pd.crosstab(vr56_sample['Import_Export'], vr56_sample['Category'])

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(vr56_contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpret the results
vr56_alpha = 0.05  # Significance level

if p < vr56_alpha:
    print("Reject the null hypothesis: There is a significant association between the two categorical variables.")
else:
    print("There is no significant association between the two categorical variables.")
```

```python
# Test of Normality {Shapiro-Wilk | Kolmogorov-Smirnov | Anderson-Darling | Jarque-Bera

def normality_tests(data):

    vr56_results = {}

    # Shapiro-Wilk test
    shapiro_statistic, shapiro_pvalue = shapiro(data)
    vr56_results["Shapiro-Wilk"] = {"statistic": shapiro_statistic, "p-value": shapiro_pvalue}

    # Kolmogorov-Smirnov test (against a normal distribution)
    ks_statistic, ks_pvalue = kstest(data, 'norm')
    vr56_results["Kolmogorov-Smirnov"] = {"statistic": ks_statistic, "p-value": ks_pvalue}

    # Anderson-Darling test
    anderson_result = anderson(data)
    vr56_results["Anderson-Darling"] = {
        "statistic": anderson_result.statistic,
        "critical_values": anderson_result.critical_values,
        "significance_level": anderson_result.significance_level
    }

    # Jarque-Bera test
    jb_statistic, jb_pvalue = jarque_bera(data)
    vr56_results["Jarque-Bera"] = {"statistic": jb_statistic, "p-value": jb_pvalue}

    return vr56_results
```

```python
from scipy import stats

# Test of Correlation (T-test)
def t_test_correlation(data1, data2):
    t_statistic, p_value = stats.ttest_ind(data1, data2)
    return {"t-statistic": t_statistic, "p-value": p_value}

for col in vr56_non_cat_vars:
    print(f"Normality tests for {col}:")
    normality_test_results = normality_tests(vr56_sample[col])
    for test_name, result in normality_test_results.items():
        print(f"  {test_name}: {result}")

    # Example of T-test
    if len(vr56_non_cat_vars) > 1:
        for i in range(len(vr56_non_cat_vars)):
            for j in range(i+1, len(vr56_non_cat_vars)):
                col1, col2 = vr56_non_cat_vars[i], vr56_non_cat_vars[j]
                t_test_result = t_test_correlation(vr56_sample[col1], vr56_sample[col2])
                print(f"T-test Correlation between {col1} and {col2}: {t_test_result}")
    print()
    print('*'*20)
    print()
```

##Time Taken


```python
import time
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
def time_model_run(model, X_train, y_train):
  """
  Measures the runtime for training a model.

  Args:
      model: The machine learning model to train.
      X_train: The training data features.
      y_train: The training data target labels.

  Returns:
      The time taken to train the model in seconds.
  """
  start_time = time.time()
  model.fit(X_train, y_train)
  end_time = time.time()
  return end_time - start_time

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55056)
log_reg_time = time_model_run(log_reg_model, vr56_X_train, vr56_y_train)
print(f"Logistic Regression runtime: {log_reg_time:.4f} seconds")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55056)
svm_time = time_model_run(svm_model, vr56_X_train, vr56_y_train)
print(f"SVM runtime: {svm_time:.4f} seconds")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55056)
sgd_time = time_model_run(sgd_model, vr56_X_train, vr56_y_train)
print(f"SGD runtime: {sgd_time:.4f} seconds")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55056)
dt_time = time_model_run(dt_model, vr56_X_train, vr56_y_train)
print(f"Decision Tree runtime: {dt_time:.4f} seconds")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_time = time_model_run(knn_model, vr56_X_train, vr56_y_train)
print(f"KNN runtime: {knn_time:.4f} seconds")

# Naive Bayes
nb_model = GaussianNB()
nb_time = time_model_run(nb_model, vr56_X_train, vr56_y_train)
print(f"Naive Bayes runtime: {nb_time:.4f} seconds")

# Random Forest
rf_model = RandomForestClassifier(random_state=55056)
rf_time = time_model_run(rf_model, vr56_X_train, vr56_y_train)
print(f"Random Forest runtime: {rf_time:.4f} seconds")

# Extreme Gradient Boosting (XGBoost)
try:
  xgb_model = XGBClassifier(random_state=55056)
  xgb_time = time_model_run(xgb_model, vr56_X_train, vr56_y_train)
  print(f"XGBoost runtime: {xgb_time:.4f} seconds")
except Exception as e:
  print("XGBoost runtime: {xgb_time:.4f} seconds")
```

Observations:

Fastest: Naive Bayes is the fastest model, taking only 0.0057 seconds to train. KNN also demonstrates very fast training times with 0.0071 seconds.
Moderately Fast: Logistic Regression (0.0183 seconds), SGD (0.0561 seconds), and Decision Tree (0.0402 seconds) have moderate training times, suitable for scenarios where faster training is desirable.
Relatively Slow: SVM (0.7826 seconds) and XGBoost (1.1844 seconds) are relatively slow compared to the others, requiring significantly more time for training.
Slowest: Random Forest is the slowest model, taking 1.5574 seconds to train. This makes it less suitable for applications where training time is a critical factor.
Insights:

Model Selection Trade-off: There is a clear trade-off between training speed and model complexity.
Faster models like Naive Bayes and KNN might offer a good starting point for initial exploration and quick analysis.
Models like Logistic Regression, SGD, and Decision Tree provide a balance between training speed and model complexity.
More complex models like SVM, XGBoost, and Random Forest often require longer training times but can potentially achieve higher accuracy.
Computational Resources: The training times of SVM, XGBoost, and Random Forest highlight the importance of considering computational resources when selecting models, especially for larger datasets or when training time is a critical factor.

#Memory Utilized

```python
import tracemalloc
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def get_model_memory_usage(model, X, y):
    tracemalloc.start()
    model.fit(X, y)
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return peak / (1024 * 1024)  # Convert bytes to MB

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55056)
log_reg_memory = get_model_memory_usage(log_reg_model, vr56_X_train, vr56_y_train)
print(f"Logistic Regression memory usage: {log_reg_memory:.2f} MB")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55056)
svm_memory = get_model_memory_usage(svm_model, vr56_X_train, vr56_y_train)
print(f"SVM memory usage: {svm_memory:.2f} MB")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55056)
sgd_memory = get_model_memory_usage(sgd_model, vr56_X_train, vr56_y_train)
print(f"SGD memory usage: {sgd_memory:.2f} MB")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55056)
dt_memory = get_model_memory_usage(dt_model, vr56_X_train, vr56_y_train)
print(f"Decision Tree memory usage: {dt_memory:.2f} MB")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_memory = get_model_memory_usage(knn_model, vr56_X_train, vr56_y_train)
print(f"KNN memory usage: {knn_memory:.2f} MB")

# Naive Bayes
nb_model = GaussianNB()
nb_memory = get_model_memory_usage(nb_model, vr56_X_train, vr56_y_train)
print(f"Naive Bayes memory usage: {nb_memory:.2f} MB")

# Random Forest
rf_model = RandomForestClassifier(random_state=55056)
rf_memory = get_model_memory_usage(rf_model, vr56_X_train, vr56_y_train)
print(f"Random Forest memory usage: {rf_memory:.2f} MB")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55056)
xgb_memory = get_model_memory_usage(xgb_model, vr56_X_train, vr56_y_train)
print(f"XGBoost memory usage: {xgb_memory:.2f} MB")

```

Observations:

Highest: XGBoost has the highest memory usage, requiring 0.88 MB, indicating a significant memory footprint during model training and execution.
Relatively High: Random Forest also has relatively high memory usage (0.45 MB), which could be a concern for applications with limited memory resources.
Moderate Memory Usage: Logistic Regression (0.34 MB), Decision Tree (0.22 MB), and KNN (0.22 MB) have moderate memory usage, making them suitable for most applications.
Low Memory Usage: SVM (0.25 MB) and SGD (0.17 MB) have relatively low memory usage, making them suitable for resource-constrained environments or applications with limited memory availability.
Lowest Memory Usage: Naive Bayes has the lowest memory usage (0.11 MB), making it ideal for memory-sensitive tasks and applications with limited memory resources.
Insights:

Memory Usage Considerations: Memory usage is an important factor to consider when selecting a model, especially in resource-constrained environments or when dealing with large datasets.
Model Choice: Models with lower memory usage, such as Naive Bayes, SGD, and SVM, are generally preferred in scenarios where memory resources are limited.
Trade-off: There is a trade-off between model complexity and memory usage. More complex models, such as XGBoost and Random Forest, often require more memory resources but can potentially achieve higher accuracy.
